```python
def engine_info(engine_text):
  pattern = r"(\d+\.\d+)([A-Z]+)|(?:\d+)([A-Z]+)"
  info= re.findall(pattern, engine_text)
  values_with_units = " ".join([match[0] for match in info])
  remaining_text = re.sub(pattern, "", engine_text).strip()
  return values_with_units, remaining_text
df_dropped[['Values with Units', 'Remaining Text']] = df_dropped['engine'].apply(engine_info).to_list()
df_dropped.head()
```

```python
def transmission_info(text):
  pattern = r"(\d+)\s([A-Z]+)|(?:\d+)([A-Z]+)"
  info= re.findall(pattern, text)
  values_with_units = " ".join([match[0] for match in info])
  remaining_text = re.sub(pattern, "", text).strip()
  return values_with_units, remaining_text
df_dropped[['gear', 'Rem_text']] = df_dropped['transmission'].apply(transmission_info).to_list()
df_dropped.head()
```

```python
import numpy as np
def remove_outliers_iqr(df, column_name):
  Q1 = df[column_name].quantile(0.25)
  Q3 = df[column_name].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return df.loc[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]

num_cols_train=df_dropped.select_dtypes(include=np.number).columns
for num in num_cols_train:
  df_dropped=remove_outliers_iqr(df_dropped,num)

df_dropped.info()
```

```python
def simplify_details(text):
  # ... (existing logic for removing terms)
  # Replace cylinder configuration terms
  text = re.sub(r' V', ' ', text)  # Replace 'v' with space
  text = re.sub(r' I', ' ', text)  # Replace 'i' with space
  text = re.sub(r' W', ' ', text)  # Replace 'w' with space
  text = re.sub(r' Straight', ' ', text)  # Replace 'straight' with space
  # Optional: Group by cylinder count (uncomment if desired)
  cylinders = re.findall(r'\d+', text)  # Extract digits using regular expressions
  if cylinders:
     text = f'{int(cylinders[0])}'

  return text.strip()

df_dropped['cylinder_type'] = df_dropped['engine_type'].apply(simplify_details)
df_dropped.head()
```

```python
def extract_injection(text):
  pattern="[A-Z]{4}I"
  text=re.sub(r' Cylinder ',' ',text)
  text=re.sub(r' Engine | engine',' ',text)
  text = re.sub(r'Straight', ' ', text)
  text=re.sub(r'([A-Z]+)(\d+)',' ',text)
  text=re.sub(r'(\d+)(\s)(\s)',' ',text)
  text=re.sub(r'(\d+)(\s)',' ',text)
  text=re.sub(r'Gas/Electric Hybrid','Hybrid Electric',text)
  text=re.sub(r'Gasoline/Mild Electric Hybrid','Hybrid Electric',text)
  text=re.sub(r'Engine Gas/Electric Hybrid','Hybrid Electric',text)
  text=re.sub(r'Plug In Electric/Gas','Plug in Hybrid',text)

  remaining_text = re.sub(pattern, "", text).strip()
  return remaining_text
df_dropped['engine_type_train'] = df_dropped['engine_type'].apply(extract_injection).to_list()
df_dropped.head()
```
